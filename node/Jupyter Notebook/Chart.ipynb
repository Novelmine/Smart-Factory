```python
data1 = maple.sort_values(by=["인기도"], ascending=False)
data1.head()
```

```python
data1.reset_index(drop=True, inplace=True) #drop False 시 예전 인덱스값 컬럼추가, inplace False 시 원본값이 변경되지 않음
data1
```

```python
data1[['유저이름','인기도']].head()
```

```python
data1 = data1.loc[0:30, ['유저이름', '인기도']]
data1
```

```python
import matplotlib.pyplot as plt
```

```python
data1.유저이름
```

```python
name = data1.유저이름.tolist()  #파이썬의 리스트 타입으로 만들어주는 함수
name
```

```python
data1.유저이름 = name
data1.유저이름.tolist()
```

```python
hr = data1.인기도.tolist()
hr
```

```python
import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':    #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
    
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
plt.figure(figsize=(10,4))
plt.plot(name, hr)
plt.show()
```

```python
plt.figure(figsize=(40,9))
plt.bar(name, hr, color='#0000ff')
plt.xlabel('유저이름')
plt.xticks(rotation=45)
plt.ylabel('인기도')
plt.title('인기도 순위')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.pie(hr, labels=name, autopct='%1.2f %%', shadow=False, startangle=-20)
plt.title('2022년 인기도 순위')
plt.show()
```
